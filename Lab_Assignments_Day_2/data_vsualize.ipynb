```python
import matplotlib.pyplot as plt
import numpy as np

#plot a line chart
x=[1,2,3,5]
y=[2,3,4,5]

plt.plot(x,y)
plt.xlabel('students')
plt.ylabel('marks')
plt.title('line chart')

#plt.show()
```

```python
#plot a bar chart with fonsize color change

x=["python","c","c++","js"]
y=[65,89,12,33]
plt.xlabel("programming language",fontsize=15)
plt.ylabel("marks",fontsize=10)
plt.title("students register",fontsize=20)

#plt.bar(x,y,color="orange")
c=['orange','green','yellow','red']
plt.bar(x,y,color=c,edgecolor="blue",label="pol")

# plt.legend()
plt.show()
```

```python
#scatter plot
x=[1,2,3,5]
y=[2,3,4,5]

c=['orange','green','yellow','red']
size=[500,400,200,100]
# plt.scatter(x,y,color=c,sizes=size)
plt.scatter(x,y,color=c,sizes=size,alpha=0.5)
plt.colorbar()
```

```python
# pie chart 

x=[14,15,20]
y=['c',"c++","c#"]

# plt.pie(x,labels=y)
# plt.pie(x,labels=y,labeldistance=0.5)
#plt.pie(x,labels=y,labeldistance=0.5,radius=2.5,autopct='%1.2f%%')

ex=[0.5,0,0]
plt.pie(x,labels=y,explode=ex,labeldistance=0.5,radius=2.5,autopct='%1.2f%%')
```
